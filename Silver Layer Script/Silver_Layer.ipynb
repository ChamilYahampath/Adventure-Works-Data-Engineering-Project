### Data Loading

#### Reading Data

```python
df_gal = spark.read.format('csv').option("header",True).option("inferSchema",True).load("abfss://bronze@weatherstoragedatalake.dfs.core.windows.net/Galle")
```

```python
df_nik = spark.read.format('csv').option("header",True).option("inferSchema",True).load("abfss://bronze@weatherstoragedatalake.dfs.core.windows.net/Nikaweratiya")
```

### TRANSFORMATIONS

#### Galle

```python
df_gal.display()
```

```python
df_gal = df_gal.withColumn("Month", month(col("Date"))) \
           .withColumn("Year", year(col("Date")))
```

```python
df_gal.display()
```

```python
df_gal.write.format('parquet')\
            .mode('append')\
            .option("path","abfss://silver@weatherstoragedatalake.dfs.core.windows.net/Galle")\
            .save()
```

#### Nikaweratiya
